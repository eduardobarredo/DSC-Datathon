**Importamos las bases de datos por año**

**2021**

```python
#Uploading the tables for 2020
path = "/content/drive/MyDrive/DATATHON_DSC/2021"
  
# csv files in the path
files = glob.glob(path + "/*.csv")
  
# defining an empty list to store 
# content
data_frame = pd.DataFrame()
content = []
  
# checking all the csv files in the 
# specified path
for filename in files:
    
    # reading content of csv file
    # content.append(filename)
    df = pd.read_csv(filename, low_memory=False, index_col=False, thousands=',')
    df = df.iloc[1: , :]
    content.append(df)
Sales2021 = pd.concat(content)
Sales2021 = Sales2021.drop(['Tipo'], axis=1)
```

**Concatenamos todos los resultados en una sola base de datos**

```python
SalesTotal = pd.concat([Sales2017,Sales2018,Sales2019,Sales2020,Sales2021])
```

**Renombramos columnas**

```python
DB_rename_columns={
SalesTotal.columns[0]: "Empresa", 
SalesTotal.columns[1]: "Region", 
SalesTotal.columns[2]: "Filial",     
SalesTotal.columns[3]: "PdVenta", 
SalesTotal.columns[4]: "KeyCliente", 
SalesTotal.columns[5]: "KeyFecha", 
SalesTotal.columns[6]: "KeyFolio", 
SalesTotal.columns[7]: "KeyTicket", 
SalesTotal.columns[8]: "Litros", 
SalesTotal.columns[9]: "Venta", 
}

SalesTotal = SalesTotal.rename(columns = DB_rename_columns)
SalesTotal['KeyFecha'] = SalesTotal['KeyFecha'].astype(int)
SalesTotal['KeyFecha'] = SalesTotal['KeyFecha'].astype(str)
SalesTotal['KeyFecha'] = pd.to_datetime(SalesTotal['KeyFecha'].astype(str))
SalesTotal['Litros'] = SalesTotal['Litros'].astype(float)
SalesTotal['Venta'] = SalesTotal['Venta'].astype(float)
```

**Ordenar por fecha**

```python
SalesTotal = SalesTotal.sort_values(by="KeyFecha")
```
